# First Unique
A non-empty array A consisting of N integers is given. The *unique number* is the number that occurs exactly once in array A.

For example, the following array A:
```python
  A[0] = 4
  A[1] = 10
  A[2] = 5
  A[3] = 4
  A[4] = 2
  A[5] = 10
```
contains two unique numbers (5 and 2).

You should find the first unique number in A. In other words, find the unique number with the lowest position in A.

For above example, 5 is in second position (because A[2] = 5) and 2 is in fourth position (because A[4] = 2). So, the first unique number is 5.

Write a function:
```python
def solution(A)
```
that, given a non-empty array A of N integers, returns the first unique number in A. The function should return −1 if there are no unique numbers in A.

For example, given:
```python
  A[0] = 1
  A[1] = 4
  A[2] = 3
  A[3] = 3
  A[4] = 1
  A[5] = 2
```

the function should return 4. There are two unique numbers (4 and 2 occur exactly once). The first one is 4 in position 1 and the second one is 2 in position 5. The function should return 4 bacause it is unique number with the lowest position.

Given array A such that:
```python
  A[0] = 6
  A[1] = 4
  A[2] = 4
  A[3] = 6
```
the function should return −1. There is no unique number in A (4 and 6 occur more than once).

Write an efficient algorithm for the following assumptions:
- N is an integer within the range [1..100,000];
- each element of array A is an integer within the range [0..1,000,000,000].


````python
# Solution 1
def solution(A:list)->int:
    unique_number = {}
    for i in A:
        if not str(i) in unique_number:
            unique_number[str(i)] = 0
        unique_number[str(i)] += 1
    
    unique = [k for k,v in unique_number.items()if v == 1]
    
    if len(unique) > 0:
        return int(unique[0])
    else:
        return -1
````

````python
# Pruebas básicas
print(solution([1, 4, 3, 3, 1, 2]) == 4)
print(solution([6, 4, 4, 6]) == -1)
print(solution([4, 10, 5, 4, 2, 10]) == 5)
# Casos con un único elemento
print(solution([7]) == 7)  # Solo un número, debe devolverlo
# Casos con todos los números repetidos
print(solution([8, 8, 8, 8]) == -1)  # Todos los números se repiten
# Casos con múltiples únicos pero uno aparece primero
print(solution([9, 10, 11, 9, 10, 12]) == 11)  # 11 es el primer número único
# Casos con los números únicos al final
print(solution([13, 13, 14, 15]) == 14)  # 14 aparece primero de los únicos
# Casos donde el primer elemento es único
print(solution([16, 17, 17, 18, 18]) == 16)  # 16 es único y aparece primero
# Casos con números grandes
print(solution([1000000000, 999999999, 1000000000]) == 999999999)  # Manejo de límites altos
# Casos con todos los números únicos
print(solution([19, 20, 21]) == 19)  # Todos son únicos, se debe devolver el primero
# Casos con entradas mixtas
print(solution([22, 23, 22, 24, 25, 24]) == 23)  # 23 es el único que no se repite
print(solution([26, 27, 28, 29, 27, 28, 26]) == 29)  # 29 es el único que no se repite
# Casos límite con tamaño máximo permitido
print(solution([i for i in range(1, 100001)] + [1]) == 2)  # Todos son únicos excepto 1
````

````python
# Solution 2
def solution(A: list) -> int:
    # Diccionario para conteo de ocurrencias
    count = {}
    # Diccionario para el primer índice de cada número
    first_index = {}
    
    # Recorremos el arreglo una vez
    for idx, num in enumerate(A):
        if num not in count:
            count[num] = 1
            first_index[num] = idx
        else:
            count[num] += 1
    
    # Buscamos el número único con menor índice
    first_unique_index = float('inf')
    result = -1
    
    for num, freq in count.items():
        if freq == 1 and first_index[num] < first_unique_index:
            first_unique_index = first_index[num]
            result = num
    
    return result
````
